#### Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

#### Import the dataset chipotle and Assign it to a variable

```python
Data = pd.read_csv("C:/Users/alan/AI Project/dataset/chipotle.csv")
myData = pd.DataFrame(Data) # use panda DataFrame re-generate data
myData
```

#### How many products cost more than $10.00?

- first we remove all the $ sign then we can convert using .astype() to convert to float
- after conversion we can compare value > 10 and find the length of table

```python
myData['item_price'] = myData['item_price'].str.replace('$', '', regex=False).astype(float)
filter = myData[myData['item_price'] > 10].reset_index()
filter
```

```python
item = len(filter)
print(str(item) + ' products are cost more than $10.00')
```

#### What is the price of each item? (print a data frame with only two columns item_name and item_price)

- we can use groupby method to groupby name and price
- The .first()  is used to get the first value of each group in a grouped dataframe. 
- When you apply first to a grouped dataframe, it returns a new dataframe with the first row of each group.
- since we only want  two columns item_name and item_price, we select [['item_name', 'item_price']]


```python
item = myData.groupby(['item_name','item_price']).first().reset_index()[['item_name', 'item_price']]
item
# new table is already in sorted order
```

#### Sort by the name of the item
when perform the groupby operation and first() method, the resulting DataFrame will be sorted based on the order of the unique values in the grouping columns

```python
# which is same as above table
item.sort_values(by='item_name', ascending=True)
```

#### What was the quantity of the most expensive item ordered?

```python
max = item.max().count()
"The quantity of the most expensive item ordered is: "+ str(max)
```

we can use len() to find the length of the table which will give us number of times

#### How many times did someone order more than one Canned Soda?

- use filter to filter Data only for Canned Soda
- convert quantity type to integer
- use boolean > 1 to filter out table with at least 2 Soda
- use len() function to get the value

```python
soda = myData[myData.item_name == "Canned Soda"]
filter = soda[soda['quantity'].astype(int) > 1]
filter
```

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

#### Create the 3 DataFrames based on the raw data in the following cell. Assign each dataframe to a variable called data1, data2, data3.

```python
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)
```

```python
data1
```

```python
data2
```

#### Join the two dataframes along rows and assign all_data.
 - Two data sets containing the same columns but different rows of data
 - use pd.concat() function

```python
all_data = pd.concat([data1,data2])
all_data
```

#### Join the two dataframes along columns and assing to all_data_col.

- reset_index(drop=True) is a method in pandas used to reset the indices of a dataframe.

```python
all_data = all_data.reset_index(drop=True)
data3 = data3.reset_index(drop=True)
all_data_col = pd.concat([all_data,data3], axis = 1)
all_data_col
```

#### Merge all_data and data3 along the subject_id value.
merge() function. similar to join function in SQL
 - we can merge(left, right, outer, inner)
      - syntax "data = pd.merge(table1, table2, how=" ")"

- left : use keys from left frame only 
- right: use key from right frame only
- outer: union of keys from both frames
- inner: intersection of keys

Leaving the join column to default in this way is not best practice. It is better to explicitly name the column using the ***on*** parameter.


```python
all_data_col = pd.merge(all_data,data3, on = "subject_id")
all_data_col
```

#### Merge only the data that has the same 'subject_id' on both data1 and data2.

```python
newData = pd.merge(data1,data2, on = "subject_id")
newData
```

#### Merge all values in data1 and data2, with matching records from both sides where available.

```python
print(data1)
print(data2)
sameData = pd.merge(data1,data2, how = "inner", on = "subject_id")
sameData

# In Python/pandas, any missing values are shown as NaN
```
